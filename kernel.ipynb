```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagen = cv2.imread("img/boy.png")
rgb= cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.title('Original')
```

```python
def espectro (kernel):
    f = np.fft.fft2(kernel)
    fshift = np.fft.fftshift(f)
    return np.log(np.abs(fshift)+1)
```

# con matriz 3x3

```python
# enfocar
kernel= np.array([[ 0,  1,  0],[ -1,  5,  -1],[0, -1, 0]], np.float32)/3

plt.subplot(121), plt.imshow(espectro(kernel),cmap = 'gray'), plt.title('núcleo'), plt.axis('off')
dst = cv2.filter2D(rgb,-1,kernel) 
plt.subplot(122), plt.imshow(dst), plt.title('convolución'), plt.axis('off')
```

```python
# Media (desenfoque)
kernel= np.array([[ 1,  1,  1],[ 1,  1,  1],[1, 1, 1]], np.float32)/9

plt.subplot(121), plt.imshow(espectro(kernel),cmap = 'gray'), plt.title('núcleo'), plt.axis('off')
plt.colorbar()
dst = cv2.filter2D(rgb,-1,kernel) 
plt.subplot(122), plt.imshow(dst), plt.title('convolución'), plt.axis('off')
```
