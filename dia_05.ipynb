```python
# 3 Realce de bordes mediante Unsharp Masking
%run script/procesador.py img/mandril_color.tif s

import cv2 as cv
import matplotlib.pyplot as plt

image= cv.imread('img/mandril_color.tif')
blur = cv.imread('result/sobel-30.png')
unsharp_image = cv.addWeighted(image, 1.5, blur, -0.5, 0, image)

plt.imshow(cv.cvtColor(unsharp_image,cv.COLOR_BGR2RGB))
```
