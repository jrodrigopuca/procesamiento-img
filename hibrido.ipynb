```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
def espectro (kernel):
    f = np.fft.fft2(kernel)
    fshift = np.fft.fftshift(f)
    return np.log(np.abs(fshift)+1)
```

```python
img1 = cv.imread('img/dog.bmp')
rgb1= cv.cvtColor(img1, cv.COLOR_BGR2RGB)
plt.subplot(121),plt.imshow(rgb1), plt.axis('off')

img2 = cv.imread('img/cat.bmp')
rgb2= cv.cvtColor(img2, cv.COLOR_BGR2RGB)

plt.subplot(122),plt.imshow(rgb2), plt.axis('off')
```

```python
dst1 = cv.blur(rgb1,(17,17))
plt.imshow(dst1), plt.axis('off')
```

```python
# laplaciano
kernel2= np.array([[ -1, 1, 1],[ -1,  -2,  1],[-1, 1, 1]], np.float32)
dst2 = cv.filter2D(rgb2,-1,kernel2) 
plt.imshow(dst2), plt.axis('off')
```

```python
#d=dst1+dst2
d2=cv.addWeighted(dst2,0.7,rgb2,0.3,0)
d1=cv.addWeighted(dst1,0.6,rgb1,0.4,0)
d3=cv.addWeighted(d1,0.5,d2,0.5,0)
plt.imshow(d3), plt.axis('off')
```
