```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
imagen = cv2.imread("boy.png")
#plt.imshow(imagen, cv2)
```

```python
# original
rgb= cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
hsv= cv2.cvtColor(imagen, cv2.COLOR_BGR2HSV)
plt.imshow(rgb)
plt.axis('off')
```

```python
# Cambiando todo h usando fill()
h,s,v = cv2.split(hsv)
h.fill(150)
nuevo_hsv=cv2.merge([h,s,v])
nuevo_rgb= cv2.cvtColor(nuevo_hsv, cv2.COLOR_HSV2RGB)
plt.imshow(nuevo_rgb)
plt.axis('off')
```

```python
# Cambiando todo h usando slice()
otro=hsv.copy()
otro[:, :, 0] = 100
plt.imshow(cv2.cvtColor(otro, cv2.COLOR_HSV2RGB))
plt.axis('off')
```

```python
# Cambiando solo un segmento de color usando where
h,s,v = cv2.split(hsv)
h=np.where((h>=100) & (h<=250),189,h)
#s=np.where((s>0)&(s<9),65,s)
#v=np.where(v==105,0,v)
nuevo_hsv=cv2.merge([h,s,v])
nuevo_rgb= cv2.cvtColor(nuevo_hsv, cv2.COLOR_HSV2RGB)
plt.imshow(nuevo_rgb)
plt.axis('off')
```

```python
# Cambiando solo un segmento de color usando where
h,s,v = cv2.split(hsv)
h=np.where((h>=27) & (h<=70),100,h)
#s=np.where((s>0)&(s<9),65,s)
#v=np.where(v==105,0,v)
nuevo_hsv=cv2.merge([h,s,v])
nuevo_rgb= cv2.cvtColor(nuevo_hsv, cv2.COLOR_HSV2RGB)
plt.imshow(nuevo_rgb)
plt.axis('off')
```
