```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
import math
```

```python
# consigna: Realice un algoritmo que procese el video 
# y retorna una imagen de la escena sin tener en cuenta 
# todos los objetos que se movieron.

# video para leer
video= cv2.VideoCapture('video/2.mp4')
w=int(video.get(cv2.CAP_PROP_FRAME_WIDTH)) #1080
h=int(video.get(cv2.CAP_PROP_FRAME_HEIGHT)) #1920

todos=np.empty((h, w))
i=0
success, frame = video.read()


# siguientes frames
while success:
    i+=1
    gris = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    todos= np.add(todos, gris)
    #probando acumulada en frame 39
    if (i==39):
        plt.imshow(todos,cmap=plt.cm.gray)
    success, frame = video.read()   
    
print("cant. de frames:", i)
```

```python
# Resultado
media= todos[1::1,1::1] / i
plt.imshow(media,cmap=plt.cm.gray)
cv2.imwrite('result/fantasmal.png', media)
```
